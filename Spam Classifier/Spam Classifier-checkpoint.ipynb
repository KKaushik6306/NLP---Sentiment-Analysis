### Text Processing

```python
def remove_pattern(input_text,pattern):
    r = re.findall(pattern,input_text)
    for word in r:
        input_text = re.sub(word,"",input_text)
    return input_text
```

```python
#Remove twitter handles (@user)
df['Message'] = np.vectorize(remove_pattern)(df['Message'],"@[\w]*")
```

```python
def l_case():
    for x in df['Message']:
        x.lowercase()
```

```python
#Change text to lower case
df['Message'] = df['Message'].apply(lambda x: x.lower()) 
```

```python
#Remove punctuation
df['Message'] = df['Message'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)))
```

```python
#Remove Special characters, numbers and other non text
df['Message'] = df['Message'] .str.replace("[^a-zA-Z#]"," ")
df.head()
```

### Remove Stop Words and perform Stemming & Lemmatization technique

```python
#List of Stop Words
stop_words = list(set(stopwords.words('english')))
```

```python
#List of Tokenized Words
tokenize_words = df['Message'].apply(lambda x: word_tokenize(x,"english"))
```

```python
#Stemming the words
stemmer = PorterStemmer()
tokenize_words = tokenize_words.apply(lambda sentence: [stemmer.stem(word) for word in sentence])


#lemmatization
lemmatizer = WordNetLemmatizer()
tokenize_words = tokenize_words.apply(lambda sentence: [lemmatizer.lemmatize(word) for word in sentence])
```

```python
#Combine tokenized words into single sentence
for i in range(len(tokenize_words)):
    tokenize_words[i] = " ".join(tokenize_words[i])
```

```python
df['New Message'] = tokenize_words
```

```python
df.head()
```

```python
#Remove Short Words
df['New Message'] = df['Message'].apply(lambda x: " ".join([w for w in word_tokenize(x,"english") if w not in stop_words]))
```

### List of Words 

```python
#List of Tokenized Words
tw = df['Message'].apply(lambda x: word_tokenize(x,"english"))
Posivite_wrds = []
Negative_wrds = []
Neutral_wrds =  []
sv = SentimentIntensityAnalyzer()
for line in tw:
    for word in line:
        score = sv.polarity_scores(word)
        neg = score['neg']
        pos = score['pos']
        if pos>neg:
            Posivite_wrds.append(word)
        elif neg > pos:
            Negative_wrds.append(word)
        else:
            Neutral_wrds.append(word)
```

```python
print(Posivite_wrds[:5])
print(Negative_wrds[:5])
print(Neutral_wrds[:5])
```

```python
def sentiment_analyses(sentiment_text):
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    neg = score['neg']
    pos = score['pos']
    if neg>pos:
        return 'Negative Sentiment'
    elif pos > neg:
        return 'Positive Sentiment'
    else:
        return 'Neutral Sentiment'
```

```python
df ['New Analysis'] = df['New Message'].apply(sentiment_analyses)
df.head()
```
